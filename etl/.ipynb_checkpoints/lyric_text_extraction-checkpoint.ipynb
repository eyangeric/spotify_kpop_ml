```python
import json
import spacy
```

```python
json_object = open('../data/track_lyrics_translated.json')
track_lyrics_translated = json.load(json_object)
```

```python
track_ids = list(track_lyrics_translated.keys())
```

```python
# words to remove: 'lyrics', ','
# look for '?', '!', '()', '-'
# remove rare words
# stop words: ‘ourselves’, ‘hers’, ‘between’, ‘yourself’, ‘but’, ‘again’, ‘there’, ‘about’, ‘once’, ‘during’, ‘out’, ‘very’, ‘having’, ‘with’, ‘they’, ‘own’, ‘an’, ‘be’, ‘some’, ‘for’, ‘do’, ‘its’, ‘yours’, ‘such’, ‘into’, ‘of’, ‘most’, ‘itself’, ‘other’, ‘off’, ‘is’, ‘s’, ‘am’, ‘or’, ‘who’, ‘as’, ‘from’, ‘him’, ‘each’, ‘the’, ‘themselves’, ‘until’, ‘below’, ‘are’, ‘we’, ‘these’, ‘your’, ‘his’, ‘through’, ‘don’, ‘nor’, ‘me’, ‘were’, ‘her’, ‘more’, ‘himself’, ‘this’, ‘down’, ‘should’, ‘our’, ‘their’, ‘while’, ‘above’, ‘both’, ‘up’, ‘to’, ‘ours’, ‘had’, ‘she’, ‘all’, ‘no’, ‘when’, ‘at’, ‘any’, ‘before’, ‘them’, ‘same’, ‘and’, ‘been’, ‘have’, ‘in’, ‘will’, ‘on’, ‘does’, ‘yourselves’, ‘then’, ‘that’, ‘because’, ‘what’, ‘over’, ‘why’, ‘so’, ‘can’, ‘did’, ‘not’, ‘now’, ‘under’, ‘he’, ‘you’, ‘herself’, ‘has’, ‘just’, ‘where’, ‘too’, ‘only’, ‘myself’, ‘which’, ‘those’, ‘i’, ‘after’, ‘few’, ‘whom’, ‘t’, ‘being’, ‘if’, ‘theirs’, ‘my’, ‘against’, ‘a’, ‘by’, ‘doing’, ‘it’, ‘how’, ‘further’, ‘was’, ‘here’, ‘than’
```

```python
feeling_indicators = {'question': '?', 'exclamation': '!', 'background_emphasis': '(', 'emphasis': '-'}

track_feeling_indicators = []
for track_id in track_ids:
    print(f"Track ID: {track_id}")
    lyric_lines = track_lyrics_translated[track_id]
    lyric_lines_lowercase = [lyric_line.lower() for lyric_line in lyric_lines]
    
    try:
        lyric_lines_lowercase.remove('lyrics')
    except:
        pass
    
    feeling_indicator_counters = []
    for lyric_line in lyric_lines_lowercase:
        feeling_indicator_counts = []
        for key, value in feeling_indicators.items():
            feeling_indicator_count = lyric_line.count(value)
            feeling_indicator_counts.append(feeling_indicator_count)
        feeling_indicator_counters.append(dict(zip(list(feeling_indicators.keys()), feeling_indicator_counts)))

    track_feeling_indicators.append({track_id: feeling_indicator_counters})   
```

```python
with open("../data/track_feeling_indicators.json", "w") as json_file:
    json.dump(track_feeling_indicators, json_file)
```

```python
characters_to_remove = [',', '.', '?', '!', '-', '(', ')', '♪']

word_counter = {}
for track_id in track_ids:
    print(f"Track ID: {track_id}")
    lyric_lines = track_lyrics_translated[track_id]
    lyric_lines_lowercase = [lyric_line.lower() for lyric_line in lyric_lines]
    
    try:
        lyric_lines_lowercase.remove('lyrics')
    except:
        pass
    
    for lyric_line in lyric_lines_lowercase:
        word_counter_keys = list(word_counter.keys())
        
        for character_to_remove in characters_to_remove:
            lyric_line = lyric_line.replace(character_to_remove, '')
            
        words = lyric_line.split(' ')
        for word in words:
            if word not in word_counter_keys:
                word_counter[word] = 0
            else:
                word_counter[word] += 1
        
```

```python
sorted(word_counter.items(), key = lambda word: word[1], reverse = True) 
```

```python
with open("../data/word_counter.json", "w") as json_file:
    json.dump(word_counter, json_file)
```

```python
words = ' '
for word, count in word_counter.items():
    word_counted = (word + ' ')*count
    words += word_counted
```

```python
feature_words = ['you', 'i', "i'm", 'me', 'my', 'oh', 'yeah', 'like', 'your', 'love', "don't", 'want',
                'know', 'now', 'no', 'baby', 'we', 'heart', 'go', 'up', 'oh']
```

```python
words_of_interest_counter = {}
for word, count in word_counter.items():
    if word in feature_words:
        words_of_interest_counter[word] = count
```

```python
with open("../data/words_of_interest_counter.json", "w") as json_file:
    json.dump(words_of_interest_counter, json_file)
```
