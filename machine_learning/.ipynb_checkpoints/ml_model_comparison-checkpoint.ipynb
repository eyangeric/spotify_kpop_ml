```python
import psycopg2
import sqlite3
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import RepeatedStratifiedKFold, RandomizedSearchCV, cross_val_score, train_test_split
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler 
from sklearn.inspection import permutation_importance
import pickle
import shap
```

```python
conn = sqlite3.connect('../data/spotify_kpop.db')
df = pd.read_sql('select * from spotify_kpop_train', conn)
```

```python
DB_NAME = "spotify_kpop_ml_models"
conn = psycopg2.connect(f"dbname={DB_NAME}")
cursor = conn.cursor()
```

```python
cursor.execute("""SELECT table_name FROM information_schema.tables
       WHERE table_schema = 'public'""")
tables = []
for table in cursor.fetchall():
    tables.append(table[0])
tables[1:]
```

```python
ml_models_sql_table = pd.read_sql('SELECT * FROM ml_models', conn)
features_sql_table = pd.read_sql('SELECT * FROM features', conn)
ml_models_evaluation_sql_table = pd.read_sql('SELECT * FROM ml_models_evaluation', conn)
importances_sql_table = pd.read_sql('SELECT * FROM importances', conn)
```

```python
ml_models_records = ml_models_sql_table.to_dict('records')
```

```python
ml_models_dict = {}
for ml_model in ml_models_records:
    model_id, model_name = ml_model.items()
    ml_models_dict[model_id[1]] = model_name[1]
```

```python
features = ['Popularity', 'energy', 'key', 'valence', 'duration_ms',
       'lyric_sentiment_score', 'lyric_word_yeah']
```

```python
le = LabelEncoder()
df["Vibe_Group_Num"] = le.fit_transform(df['Vibe_Group'])
```

```python
outcome = 'Vibe_Group_Num'
```

```python
features_outcome = features.copy()
features_outcome.append(outcome)
```

```python
df_no_na = df[features_outcome].dropna()
```

```python
TEST_SIZE = 0.2
RANDOM_STATE = 24
EVAL_METRIC = 'logloss'
FOLDS = 10
ITERATIONS = 3
```

```python
X = df_no_na[features]
y = df_no_na[[outcome]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = TEST_SIZE, random_state = RANDOM_STATE, stratify = y)
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X_train)
scaled_features_df = pd.DataFrame(scaled_features, index=X_train.index, columns=X_train.columns)

pickle.dump(scaler, open(f"standardscaler.pkl", 'wb'))
```

```python
ss_imported = pickle.load(open(f"standardscaler.pkl", 'rb'))
```

```python
X_test_scaled = ss_imported.fit_transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, index=X_test.index, columns=X_test.columns)
```

```python
knn_log_losses = {}
for k in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, scaled_features_df, y_train, scoring = 'neg_log_loss', cv = FOLDS, n_jobs = -1, error_score = 'raise')
    logloss = -np.mean(scores)
    knn_log_losses[k] = logloss
best_knn_neighbors = sorted(knn_log_losses.items(), key = lambda k: k[1])[0][0]    
```

```python
xgb = XGBClassifier(objective = 'multi:softmax')
lr = LogisticRegression(multi_class='multinomial')
knn = KNeighborsClassifier(n_neighbors = best_knn_neighbors)
svm = SVC(probability=True)
rf = RandomForestClassifier()

models = {'XGBoost': xgb, 'Random Forest': rf, 'Support Vector Machine': svm, 'K-Nearest Neighbors': knn, 'Logistic Regression': lr}
```

```python
for model_name, model in models.items():
    print(model_name)
    print('')
    
    model_index = list(ml_models_dict.values()).index(model_name)
    model_id = list(ml_models_dict.keys())[model_index]
    hyperparameters = str(model.get_params())

    scores = cross_val_score(model, scaled_features_df, y_train, scoring = 'neg_log_loss', cv = FOLDS, n_jobs = -1, error_score = 'raise')
    eval_metric_scores_avg = np.mean(scores)
    model.fit(scaled_features_df, y_train)
    predictions = model.predict(X_test_scaled)
    outcome_df = pd.DataFrame({'Actual': y_test[outcome].tolist(), 'Predicted': predictions})
    
    dz_precision_recall_num = len(outcome_df[(outcome_df['Actual'] == 0) & (outcome_df['Predicted'] == 0)])
    dz_precision_denom = len(outcome_df[outcome_df['Predicted'] == 0])
    dz_recall_denom = len(outcome_df[outcome_df['Actual'] == 0])
    dz_precision = round(dz_precision_recall_num/dz_precision_denom*100, 1)
    dz_recall = round(dz_precision_recall_num/dz_recall_denom*100, 1)
    
    ey_precision_recall_num = len(outcome_df[(outcome_df['Actual'] == 1) & (outcome_df['Predicted'] == 1)])
    ey_precision_denom = len(outcome_df[outcome_df['Predicted'] == 1])
    ey_recall_denom = len(outcome_df[outcome_df['Actual'] == 1])
    ey_precision = round(ey_precision_recall_num/ey_precision_denom*100, 1)
    ey_recall = round(ey_precision_recall_num/ey_recall_denom*100, 1)
    
    ey_dz_precision_recall_num = len(outcome_df[(outcome_df['Actual'] == 2) & (outcome_df['Predicted'] == 2)])
    ey_dz_precision_denom = len(outcome_df[outcome_df['Predicted'] == 2])
    ey_dz_recall_denom = len(outcome_df[outcome_df['Actual'] == 2])  
    ey_dz_precision = round(ey_dz_precision_recall_num/ey_dz_precision_denom*100, 1)
    ey_dz_recall = round(ey_dz_precision_recall_num/ey_dz_recall_denom*100, 1)
    
    accuracy_num = dz_precision_recall_num + ey_precision_recall_num + ey_dz_precision_recall_num
    accuracy_denom = len(outcome_df)
    overall_accuracy = round(accuracy_num/accuracy_denom*100, 1)
    
    print(f"Training Model Log Loss: {-eval_metric_scores_avg}")
    print(f"Overall Accuracy: {overall_accuracy}%")
    print(f"Among the actual DZ's songs, {dz_recall}% were predicted correctly.")
    print(f"Among the predicted DZ's songs, {dz_precision}% were actually correct.")
    print(f"Among the actual EY's songs, {ey_recall}% were predicted correctly.")
    print(f"Among the predicted EY's songs, {ey_precision}% were actually correct.")
    print(f"Among the actual EY and DZ's songs, {ey_dz_recall}% were predicted correctly.")
    print(f"Among the predicted EY and DZ's songs, {ey_dz_precision}% were actually correct.")
    print('')
    
    try:
        feature_importance = pd.DataFrame({"Feature": X_test_scaled.columns[0:len(X_test_scaled.columns)], "Feature_Importance": model.feature_importances_})
    except:
        feature_importance = pd.DataFrame({"Feature": X_test_scaled.columns[0:len(X_test_scaled.columns)], "Feature_Importance": None})
    permutation_importance_results = permutation_importance(model, X_test_scaled, outcome_df['Actual'], n_repeats = ITERATIONS, random_state = RANDOM_STATE)
    permutation_importance_df = pd.DataFrame({"Feature": X_test_scaled.columns[0:len(X_test_scaled.columns)],
                                         'Permutation_Importance': permutation_importance_results['importances_mean']}) 
    final_importance_df = pd.merge(feature_importance, permutation_importance_df, on = 'Feature')
    final_importance_df = final_importance_df.sort_values(by=['Permutation_Importance'], ascending = False)    
    final_importance_df = pd.merge(final_importance_df, features_sql_table, left_on = 'Feature', right_on = 'name')
    final_importance_df = final_importance_df[['id', 'Feature', 'Feature_Importance', 'Permutation_Importance']]
    print(final_importance_df)
    print('')
                                                          
    try:
        explainer = shap.Explainer(model)
        shap_values = explainer.shap_values(X_test_scaled,approximate=True)
        shap.summary_plot(shap_values, X_test_scaled, plot_type="bar")
    except:
        pass
    
    print('')
    print('')
    
#     SQL = 'INSERT INTO ml_models_evaluation (ml_model_id, hyperparameters, log_loss, dz_precision, dz_accuracy, ey_precision, ey_accuracy, ey_dz_precision, ey_dz_accuracy, overall_accuracy) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);'
#     DATA = (model_id, hyperparameters, -eval_metric_scores_avg, dz_precision_recall_num/dz_precision_denom, dz_precision_recall_num/dz_recall_denom, ey_precision_recall_num/ey_precision_denom, ey_precision_recall_num/ey_recall_denom, ey_dz_precision_recall_num/ey_dz_precision_denom, ey_dz_precision_recall_num/ey_dz_recall_denom, accuracy_num/accuracy_denom)
#     cursor.execute(SQL, DATA)
#     conn.commit()
#     model_train_id = pd.read_sql('SELECT model_train_id FROM ml_models_evaluation', conn)['model_train_id'].max()
#     model_train_id = int(model_train_id)
    
#     importance_dict = final_importance_df.to_dict('records')
    
#     importance_sql = 'INSERT INTO importances (model_train_id, feature_id, feature_importance, permutation_importance) VALUES (%s, %s, %s, %s);'
#     for feature in importance_dict:
#         DATA = (model_train_id, feature['id'], feature['Feature_Importance'], feature['Permutation_Importance'])
#         cursor.execute(importance_sql, DATA)
#         conn.commit()
        
    
    
```
