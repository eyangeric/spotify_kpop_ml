```python
import json
from googletrans import Translator
```

```python
translator = Translator(service_urls = ['translate.google.co.kr'])
```

```python
json_object = open('../data/track_lyrics.json')
track_lyrics = json.load(json_object)
```

```python
def translate_lyric_line(LYRIC_LINE):
    translation = translator.translate(LYRIC_LINE, dest = 'en')
    return translation.text
```

```python
def detect_language(LYRIC_LINE):
    predicted_language = translator.detect(LYRIC_LINE)
    return predicted_language.lang
```

```python
track_id_lyric_lines = {}
for track_id, lyrics in list(track_lyrics.items()):
    lyric_lines = lyrics.split('\n')
    
    print(f"Track ID: {track_id}")
    track_lyric_lines = []
    for lyric_line in lyric_lines:
        print(f"Original Lyric Line: {lyric_line}")

        try:
            translation = translate_lyric_line(lyric_line)
            print(f"Translated Lyric Line: {translation}")
            print('')
            track_lyric_lines.append(translation)
        except:
            track_lyric_lines.append(lyric_line)
            print('')
            
        
    track_id_lyric_lines[track_id] = track_lyric_lines
```

```python
with open("../data/track_lyrics_translated.json", "w") as json_file:
    json.dump(track_id_lyric_lines, json_file)
```
