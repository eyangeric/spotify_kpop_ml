```python
import json
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
json_object = open('../data/word_counter.json')
word_counter = json.load(json_object)
```

```python
words = ' '
for word, count in word_counter.items():
    word_counted = (word + ' ')*count
    words += word_counted
```

```python
wordcloud = WordCloud(collocations=False, width = 800, height = 800,
                background_color ='white',
#                 stopwords = ['a'],
                min_font_size = 10).generate(words)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
feature_words = ['you', 'i', "i'm", 'me', 'my', 'oh', 'yeah', 'like', 'your', 'love', "don't", 'want',
                'know', 'now', 'no', 'baby', 'we', 'heart', 'go', 'up', 'oh']
```

```python
words_of_interest_counter = {}
for word, count in word_counter.items():
    if word in feature_words:
        words_of_interest_counter[word] = count
```

```python
sorted(words_of_interest_counter.items(), key = lambda word:word[1], reverse = True)
```
